```python
import torch
import make_data
import run_models
```

```python
def get_target_function(x):
        # True function is x * sin(x + 5)
        return x*torch.sin(x+5)
x, y = make_data.make_initial_data(get_target_function)
x_test, y_test, y_perfect = make_data.make_test_data(get_target_function)
```

```python
reused_params = {"x": x, "y": y, "x_test": x_test, "y_test": y_test, "y_perfect": y_perfect, "function_type": "xsin(x+5)",}
```

```python
param_set = [
    {"model_type": "bnn_variational", "dropout_prob": 0, "epochs": 4000, "samples": 100, "width": 1, "depth": 1},
]
for set in param_set:
    set.update(reused_params)
```

```python
for set in param_set:
    run_models.run_model(**set)
```
